# Replicating Pixels

```python
def replicatingPixel(img):
    
    # Assuming img input have nxnx3 dimension ie. same width and hight
    img_size = img.shape[0]
    enlar_size = img_size*img_size
    
    enlar = np.empty([img_size*img_size,img_size*img_size,3],dtype = np.uint8,order = 'C')
    
    for i in range(enlar_size):
        for j in range(enlar_size):
            
            x = i // img_size
            y = j // img_size
            enlar[i][j] = img[x][y]
            
    return enlar
```

```python
# Generate a row
def generateRow(data,cnt):
    n = 20
    row = []
    i = 0
    while i < n-1:
        baseimg = cv2.imread('./image/' + data[cnt])

        if len(row) == 0:
            row = baseimg
            continue
        row = np.hstack((row,baseimg))
        i += 1
        cnt += 1
    return row,cnt

# Adding rows vertically to generate a img
def generateimg(data,cnt):
    n = 20
    col = []
    i = 0
    while i < n-1:
        baserow,cnt = generateRow(data,cnt)

        if len(col) == 0:
            col = baserow
            continue
        col = np.vstack((col,baserow))
        i += 1
    return col

# Function to generate unique images
def getUniqueImage(data):
    
    if len(data) < 400:
        data.extend(data[0:400 - len(data)])
    
    img = generateimg(data,0)
    
    return img
    
data = []
for file in os.listdir('./image'):
    data.append(file)   
```
